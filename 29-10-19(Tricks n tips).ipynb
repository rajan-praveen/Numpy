## Histogram

```python
import matplotlib.pyplot as plt

a,b=2,0.5
c=np.random.normal(a,b,10000)  # a to b dicende into 10000    # normal distribution
print(c.shape)

plt.hist(c,bins=1000,density=10)    # 
plt.show()
```

```python

plt.hist(c,bins=1000,density=100)    # data is shrink bcoz denstiy increased
plt.show()
```

```python
plt.hist(c,bins=100,density=10,color='y')    # bins no of data u want tinto graph
plt.show()
```

```python
(n,bins)=np.histogram(c,bins=50,density=True)
plt.plot(.5*(bins[1:]+bins[:-1]),n)
plt.show()
```
